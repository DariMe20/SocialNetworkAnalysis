# **Analyzing Urban Traffic Networks using SNA: Cluj Napoca road system**

## Install and import libraries
Install osmnx (to get real city graphs) and import common SNA tools.

```python
# --- Imports ---
import osmnx as ox
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import pearsonr

# Configure OSMnx settings (smaller figures, white background)
ox.settings.log_console = True
ox.settings.use_cache = True
ox.settings.default_graph_boundary = "city"
```

## Build Road Graph for Cluj Napoca

osmnx downloads OpenStreetMap data and builds a directed weighted graph of roads.

Nodes = intersections, Edges = drivable road segments.

**Cluj napoca raw road network**

```python
# Download Cluj-Napoca road network (driveable streets)
city = "Cluj-Napoca, Romania"
G_raw = ox.graph_from_place(city, network_type="drive")

print("Raw graph:", len(G_raw.nodes()), "nodes,", len(G_raw.edges()), "edges")

# Plot non-simplified network
fig, ax = ox.plot_graph(G_raw, node_size=3, edge_color="gray", bgcolor="white")
```

**Cluj Napoca simplified network**

```python
# Simplify graph (merge intermediate points along same street)
G = G_raw

print("Simplified graph:", len(G.nodes()), "nodes,", len(G.edges()), "edges")

# Plot simplified version
fig, ax = ox.plot_graph(G, node_size=5, edge_color="steelblue", bgcolor="white")
```

## Download and visualize road network

```python
# --- 1. Create road graph for Cluj-Napoca (or any city you want) ---
city = "Cluj-Napoca, Romania"
G = ox.graph_from_place(city, network_type="drive")

# --- 2. Simplify and project coordinates for plotting ---
G = ox.simplify_graph(G)

# --- 3. Basic visualization ---
fig, ax = ox.plot_graph(G, node_size=5, bgcolor="white")
```
